### **COVID-19 Analysis**

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df=pd.read_csv(url)
df.head(10)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
loc_india=df[df['location']=='India'] #find the data for India only
loc_india[['date','new_cases']].tail() #to track new cases'''
loc_india
```

```python
date=pd.to_datetime(loc_india['date'])
date
month=loc_india.groupby(date.dt.strftime('%B'))# %B-Month, %d-Day, %y-Year, %r-Time
month
```

```python
loc_india=df[df['location']=='India'] #find the data for India only
loc_india2=loc_india.groupby(date.dt.strftime('%Y')).sum()# %B-Month, %d-Day, %y-Year, %r-Time
loc_india2
```

```python
plt.rcParams["figure.figsize"] = (10,15) # to resize the chart
plt.bar(loc_india['date'],loc_india['new_cases'])
plt.title('India New Cases')
```

### Top 10 Country for Covid-19 Cases

```python
loc_all=df[~df['iso_code'].str.contains('OWID_')] # (~)--> used to remove the row contains 'OWID_'
top10_new_cases=loc_all.groupby('location').sum()
a=top10_new_cases['new_cases'].sort_values(ascending=False).head(10)
a.plot(kind='pie')
```

### Top 10 Country for death cases Covid-19

```python
top10_new_death=loc_all.groupby('location').sum()
a=top10_new_death['new_deaths'].sort_values(ascending=False).head(10)
a.plot(kind='bar')
```

### Month-On-Month comaprison India

```python
#April'20 India data New Cases
startdt=loc_india[loc_india['date']<='2020-04-30']
monthly20=startdt[startdt['date']>='2020-04-01']
apr20=monthly20['new_cases'].sum()

#May'21 India data New Cases
startdt=loc_india[loc_india['date']<='2021-04-30']
monthly21=startdt[startdt['date']>='2021-04-01']
apr21=monthly21['new_cases'].sum()

#May'20 India data New Cases
startdt=loc_india[loc_india['date']<='2020-05-31']
monthly20=startdt[startdt['date']>='2020-05-01']
may20=monthly20['new_cases'].sum()

#May'21 India data New Cases
startdt=loc_india[loc_india['date']<='2021-05-31']
monthly21=startdt[startdt['date']>='2021-05-01']
may21=monthly21['new_cases'].sum()
apr20,may20,apr21,may21
```

```python
plt.bar("April'20",apr20,width=0.2,color='green')
plt.bar("May'20",may20,width=0.2,color='green')
plt.bar("April'21",apr21,width=0.2,color='red')
plt.bar("May'21",may21,width=0.2,color='red')
```
