```python
import pandas as pd
import datetime
```

```python
url='https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df=pd.read_csv(url)
df.head()
```

```python
lct=df.groupby('continent').sum()
lct[['total_cases','new_cases','new_deaths']]
```

```python
lct.describe()
```

```python
df1=df[df['location']=='India']
df1
df2=df1[df1['date']>='2021-04-01']
df2
```

```python
df2[['date','total_cases','total_deaths','new_cases','new_deaths']]
```

```python
date=df2['date']
New_Cases=df2['new_cases']
death=df2['new_deaths']
print(date,New_Cases,death)
```

```python
import matplotlib.pyplot as plt
plt.figure(True)
plt.rcParams["figure.figsize"] = (10,5) # to resize the chart
#plt.plot(date,New_Cases,'ro-', label='line 1', linewidth=1)
plt.plot(date,death,'ro-', label='line 1', linewidth=1)
plt.title('India New Cases')
plt.ylabel('# of Cases')
plt.xlabel('Date of Cases')
plt.show
```
