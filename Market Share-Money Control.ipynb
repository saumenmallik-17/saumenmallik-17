```python
import pandas as pd
data=pd.read_html('https://www.moneycontrol.com/broker-research/',skiprows=0)#reading data from HTML in pandas
stock=data.pop(1) #pop used to remove 1 rows from html
stock.columns=['Stocks','Buy','Hold','Sell','CMP','Median','Low','High']#renaming the columns head
print("Money Control Recommendation")#Printing Heading
stock #Printing Data
```

```python
# Top Recomended stock to buy
Top_Recomended=stock.sort_values(by=['Buy'],ascending=False)
Top_Recomended.head(10)
```

```python
from nsetools import Nse
nse=Nse()
data=nse.get_stock_codes()
stocksym=pd.DataFrame.from_dict(data,orient='index')
stocksym.columns=stocksym.iloc[0] #define the first line of data as a header
stocksym.drop('SYMBOL',axis=0,inplace=True)#remove the first line of dataframe
stocksym.head(10)
```

```python
from nsepy import get_history
from datetime import date
stockhistory=get_history(symbol='HDFCBANK',start=date(2021,5,1),end=date.today())
#HDFCBANK['Close'].plot(kind='bar')
stockhistory
```

```python
OPNCLS=stockhistory['Open']-stockhistory['Close']
HIGLOW=stockhistory['High']-stockhistory['Low']
OPNCLS.plot(kind='line')
HIGLOW.plot(kind='line')
```
